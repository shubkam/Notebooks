```python
# dictionary can be converted to dataframes with the DataFrame() function
# keys become col headers and values becomes the corresponding cols
import pandas
d  = {"Id" : [1,2,3], "Name" : ["A", "B", "C"], "Age" : [25, 26, 26]}
df = pandas.DataFrame(d)
df
```

```python
df["Age"].unique() # display only unique values from a data series
```

```python
df[["Age"]] >= 26 # displays a dataframe of cells of the column that meet the condition
```

```python
df[df['Age'] >= 26] # displays a dataframe of rows that meet the condition for that column
```

```python
# casting a list of lists to a dataframe
A = [[1,2,2], [3,4,5], [4,5,6]]
df = pandas.DataFrame(A)
df
```
